```python
%pylab inline
#from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import numpy as np
import string
import matplotlib.cm as cm
import os
from scipy import ndimage
import pandas as pd
```

```python

rll=0.139 # Radio geométrico de las llantas en traseras (m)
refc=0.98*rll # Radio efectivo llantas traseras (m)
rin=0.0631 #radio interno del neumático (m)
rlld=0.129 # Radio llantas delanteras (m)
rind=0.0631 # Radio rin llantas delanteras (m)

# Datos Transmisión
dp=11 # Dientes del piñon
dpl=39 # Dientes del plato
ng=dpl/dp # Relación de Transmisión
etaf=0.95 # Eficiencia de transmisión

# Datos eje llantas
reje=0.02 # Radio eje de las llantas (m)
meje=3.832 # Masa del eje (kg)

# Datos Kart general
masa=152.0 # masa del Kart (kg)
g=9.81 # Gravedad (m/s)
w=masa*g # peso del Kart (N)

# Datos Fuerzas de la fórmula de aceleración
rho=1.1955 # densidad del aire (kg/m^3)
dll=0.975 # Distancia entre ejes de llantas delanteras (m)
alj=0.88 # Altura con el piloto sentado en el kart (estatura 1.68 m) (m)
af=(dll*alj)/2 # Area Frontal Kart (m^2)
cd=0.89 # Coeficiente de Arrastre Kart
vw=0.0 # velocidad del viento (m/s)
theta=0.0 # Angulo carretera

# Inercias
ii=5.2E-3 # Inercia motor de combustión (kg m^2) 

mllt=1.43 # Masa llantas traseras (kg)
mlld=1.20 # Masa llantas delanteras (kg)

mbt=0.7*mllt # El porcentaje de la masa total que pesa la parte "banda" de la llanta trasera
mdt=0.3*mllt # El porcentaje de la masa total que pesa la parte "disco" de la llanta trasera
mbd=0.7*mlld # El porcentaje de la masa total que pesa la parte "banda" de la llanta delantera
mdd=0.3*mlld # El porcentaje de la masa total que pesa la parte "disco" de la llanta delantera
illd=(mbd*rlld)+(mdd*(rlld**2-rind**2))/2 # Inercia 1 llanta delantera (kg-m^2)
illt=(mbt*rll)+(mdt*(rll**2-rin**2))/2 # Inercia 1 llanta trasera (kg-m^2)

#paso=7.62 #(mm)
#rpim= paso*(0.6+(1/tan(180/dp))) # Radio Piñon (mm)
#rplam=paso*(0.6+(1/tan(180/dpl))) # Radio Plato (mm)
#rpi= rpim/1000 #Radio Piñon (m)
#rpla= rplam/1000 #Radio Plato (m)
rpi=0.01525#(m)
rpla=0.04096 #(m)
rhoac=7850.0 # Densidad del Acero (kg/m^3)
bpi=0.01 # Ancho piñon (m)
bpla=0.02 # Ancho plato (m)
vpi= bpi*np.pi*rpi**2 # Volumen del piñon (m^3)
vpla=bpla*np.pi*rpla**2 # Volumen del plato (m^3)
mpi=rhoac*vpi # Masa del piñon (kg)
mpla=rhoac*vpla # Masa del plato (kg)
ig=((mpi*(rpi**2))/2.0)+((mpla*(rpla**2))/2.0) # Inercia de la transmisión (kg-m^2)
 

ieje=(meje*reje**2)/2.0 # Inercia del eje de las llantas (kg-m^2)

# Fuerzas Aceleración
# Resistencia a la rodadura
a=0.2231 # Distancia del centro de gravedad al eje de la llanta trasera (m)
l=1.0063 # Distancia de eje de las llantas traseras y delanteras (m)
h=0.07478 # Altura del centro de gravedad del kart con respecto al piso (m)

gamma = 1 # aceleracion

# Aceleración, Fuerza de tracción
mr=((((ii-ig)*ng**2))+ieje+illd*2+illt*2)/(refc**2)
```

<div style="text-align: justify"> 1. Despues de agregar las librerias y las constantes del carro se realiza la funcion que lea archivos de VBOX, esta funcion sirve para los 2 tipos de archivo vbox con los que cuenta la universidad. La funcion tiene como entrada solo 2 cosas: la carpeta donde se encuentra el archivo y el nombre del archivo. Para este caso la carpeta solo tiene 2 opciones: data para datos de carro de combustion interna y data electrico para datos de carro electrico.</div> 

```python
def datosVBOX (base_dir,data_file):

    #base_dir = 'data_electrico'
    with open(os.path.join(base_dir,data_file), 'r', encoding='cp1252') as vboxdata:
        section = []
        time = []
        lat = []
        long = []
        vel = []
        height = []
        vertv = []
        for line in vboxdata:
            if line[0] == '[':
                section = str.replace(line, "\n","")
                continue
            if section == '[column names]':
                col_names = str.replace(line, "  \n","").split('  ')
                section = []
            if section == '[data]':
                linesplit = str.replace(line, " \n","").split(' ')
                time.append(float(linesplit[1][0:2])*3600+float(linesplit[1][2:4])*60+float(linesplit[1][4:]))
                lat.append(float(linesplit[2]))
                long.append(float(linesplit[3]))
                vel.append(float(linesplit[4]))
                height.append(float(linesplit[6]))
                vertv.append(float(linesplit[7]))

    time = np.asarray(time)
    lat = np.asarray(lat)
    long = np.asarray(long)
    vel = np.asarray(vel)*(1000/3600)
    height = np.asarray(height)
    vertv = np.asarray(vertv)
    time_base = time[0]
    time = time - time_base
    lat = lat/60.0
    long = -long/60.0
    dt = time[1]-time[0]
    datos = {"Tiempo":time, "Latitud":lat, "Longitud":long,"Velocidad":vel,"Altura":height, "dt":dt}
    tabladatos= pd.DataFrame(datos)
    return tabladatos
```

<div style="text-align: justify"> 2. En segunda instanca se definiran las funciones necesarias para Comparar el rendimiento del vehículo de combustión interna (datos Córdoba) con el del kart eléctrico (datos adjuntos) en términos de tiempo de vuelta, velocidades instantáneas, aceleraciones. Se definira esta seccion en 3 partes: La primera es una funcion de cortar datos, para analizar solo una parte de la lista de datos disponible debido a que hay datos que no sirven o para analizar solo una vuelta en cada caso. La segunda parte es la definicion de fuerzas y aceleraciones y la definicion de la funcion acelyvel, la cual retorna los la distancia recorrida en terminos de latitud y longitud, la grafica de velocidad contra tiempo, aceleracion contra tiempo y tiempo de vuelta del proceso. Por ultimo, la tercer parte se centra en la funcion TyP la cual se centra en las caracteristicas principales del motor del kart en terminos de torque, velocidad angular y potencia. </div>

```python
def cortardatos(basedir,data,i0,i1):

    x=datosVBOX(basedir,data)
    
    timef = x['Tiempo'][i0:i1]-x['Tiempo'][i0]
    latf = x['Latitud'][i0:i1]
    longf = x['Longitud'][i0:i1]
    velf = x['Velocidad'][i0:i1]
    heightf = x['Altura'][i0:i1]
    dt=x['Tiempo'][1]-x['Tiempo'][0]
    datosf = {"Tiempo":timef, "Latitud":latf, "Longitud":longf,"Velocidad":velf,"Altura":heightf, "dt":dt }
    
    
    return datosf
```

<div style="text-align: justify">Para la funcion aceleracion, se utilizara una funcion de fitro gaussiano la cual tiene como entrada la velocidad, el delta de tiempo o intervalo de tiempo de un dato a otro y el filtro, entre mayor el numero del filtro mayor se suaviza el ruido de la funcion para encontrar la aceleracion. Las demas funciones como velocidad angular, potencia y torque ya fueron definidas en el taller anterior con el mismo modelo dinamico. </div>

```python
def accel(vx, ddt,sigma):
    return ndimage.gaussian_filter1d(vx, sigma, order=1, mode='wrap') / ddt
```

```python
def omega(vx):
    x=(vx*ng)/refc

    return x # (rad/s)



def potencia(vx):
    o=omega(vx)
    omega1=o*(60/(2*pi))
    P=torque(vx)*omega1

    return P

def torque(vx, ax):
    vk=vx*(3600/1000)
    
    T=((((ax)*(masa+mr))+frc(vk)+farrastre(vx))*(1/ng*etaf)*refc)
    return T

def frodadura(vx):
    wr=w*((a/l)+(ax/g*(h/l))) # fuerza vertical sobre las llantas tractivas
    fr=0.01*(1+(vx/100.0)) # Factor de rodadura la velocidad del kart debe estar en (km/h)
    # fr=C*((mllt*g)/(rllt*2))*sqrt(ht/bllt)
    # ht=0.0739 # Altura del neumático (m)
    # bllt=0.209 # Ancho del neumático (m)
    # C= # Constante del material de la llanta
    return wr*fr # (N)

def frc(vx):
    vk=vx*(3600/1000)
    fr=0.01*(1+(vx/100.0)) # Factor de rodadura la velocidad del kart debe estar en (km/h)
    return w*fr # (N)

def farrastre(vx):
    return (0.5)*rho*af*cd*(vx)**2 # (N)

def ftraccion(ax,Fg,Fw,Fr):
    
    return masa*ax+Fg+Fw+Fr

```

<div style="text-align: justify"> La funcion acelyvel es una de las mas importantes de esta seccion, la funcion de esta funcion es definir todas las caracteristicas de Kart, en terminos de la ruta recorrida, la grafica de velocidad como funcion del tiempo, la aceleracion como funcion del tiempo, la distancia de todo el recorrido, la velocidad maxima y la aceleracion como entrada. La entrada de la funcion es la carpeta y nombre del archivo ya que utiliza la funcion cortar datos; tambien le entran 2 numero i0 e i1 como el numero de inicio y final de la lista y por ultimo electricoocombustion es un numero entre 1 y 2 (1 si los datos son electricos y 2 si son de combustion) debido a que de esta forma el filtro gaussiano para la aceleracion cambia. El filtro es mas fuerte para los datos de combustion. El valor de este filtro fue elegido de maner arbitaria, cordoba utilizo un valor de 20 para los datos de combustion tambien. </div>

```python
def acelyvel(carpeta,base,i0,i1,electricocombustion):
    
    #para electricocombustion= 1 es electrico y 2 es mecanico
    c=cortardatos(carpeta,base,i0,i1)
    figure(figsize(15,5))
    plt.subplot(1,3,1 )
    plot(c['Longitud'],c['Latitud'])
    plt.xlabel('Longitud (s)')
    plt.ylabel('Latitud (m/s)')
    plt.grid()
    
    plt.subplot(1,3,2)
    plot(c['Tiempo'],c['Velocidad'])
    plt.xlabel('t (s)')
    plt.ylabel('Velocidad(m/s)')
    plt.grid()
    
    
    plt.subplot(1,3,3)
    
    if electricocombustion==1:
        a=accel(c['Velocidad'],c['dt'],12 )
        plt.plot(c['Tiempo'],a)
        plt.xlabel('t (s)')
        plt.ylabel('Aceleracion(m/s^2)')
        plt.grid()
        plt.show()
    
    else:
        a=accel(c['Velocidad'],c['dt'],20 )
        plt.plot(c['Tiempo'],a)
        plt.plot(c['Tiempo'],a)
        plt.xlabel('t (s)')
        plt.ylabel('Aceleracion(m/s^2)')
        plt.grid()
        
    
    tiempodevuelta=np.max(c['Tiempo'])
    mensaje='el tiempo de vuelta es',tiempodevuelta
    vmax=np.max(c['Velocidad'])
    mensaje1='la velocidad maxima es',vmax
    amax=np.max(a)
    mensaje2='la aceleracion instantanea maxima es igual a',amax
    
    return mensaje,mensaje1,mensaje2
```

Por ultimo, la funcion TyP. Esta funcion tambien es una de las mas importantes del taller. Esta funcion tiene como entrada las mismas variables que la funcion acelyvel (carpeta y nombre del arhivo, el numero inicial y final de datos y el valor 1 y 2 para electroocombustion). La funcon retora la grafica de torque contra tiempo, velocidad contra tiempo, potencia contra tiempo y torque contra velocidad angular.

```python
def TyP(carpeta,base,i0,i1,electricocombustion):
    #para electricocombustion= 1 es electrico y 2 es mecanico
    c=cortardatos(carpeta,base,i0,i1)
    o=omega(c['Velocidad'])
    
    
    if electricocombustion==1:
        a=accel(c['Velocidad'],c['dt'],10)
        t=torque(c['Velocidad'],a)
    
    
    else:
        a=accel(c['Velocidad'],c['dt'],20)
        t=torque(c['Velocidad'],a)
    
    
    P=t*o
    
    figure(figsize(19,14))
    plt.subplot(2,3,1)
    plot(c['Tiempo'],t)
    plt.xlabel('t (s)')
    plt.ylabel('Torque(N*m)')
    plt.grid()
    plt.subplot(2,3,2)
    plot(c['Tiempo'],o)
    plt.xlabel('t (s)')
    plt.ylabel('Velicidad angular(rad/s)')
    plt.grid()
    plt.subplot(2,3,3)
    plot(c['Tiempo'],P)
    plt.xlabel('t (s)')
    plt.ylabel('Potencia (W)')
    plt.grid()
    plt.subplot(2,3,(4,6))
    plot(o,t)
    plt.xlabel('Velocidad angular(rad/s)')
    plt.ylabel('Torque (N*m)')
    plt.grid()
    tt=np.max(t)
    mensaje=' El Torque maximo es igual a',tt
    oo=np.max(o)
    mensaje1='La velocidad maxima es igual a',oo
    pp=np.max(P)
    mensaje2='La Potencia maxima entregada por el motor es igual a',pp
    
    return mensaje,mensaje1,mensaje2
```
